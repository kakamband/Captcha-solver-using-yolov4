```python
import os
from yolov4_detection import detect
from PIL import Image
import matplotlib.pyplot as plt
import cv2

YOLO_DATA_DIR = 'yolo-data'
TEST_DATA_DIR = 'test-treated-4'
filenames = os.listdir(TEST_DATA_DIR)
MIN_CONFIDENCE  = 0.5
NON_MAXIMA_THRESHOLD = 0.3 

results = []
for name in filenames:
    image_path = os.path.join(TEST_DATA_DIR, name)
    result, _ = detect(image_path, YOLO_DATA_DIR, MIN_CONFIDENCE, NON_MAXIMA_THRESHOLD)
    result = [value[0] for value in result]
    results.append("".join(result))
```

```python
total_test_images = len(filenames)
fig = plt.figure(figsize=(15,20))

for i in range(total_test_images):
    subplot = fig.add_subplot(total_test_images//4+1, 4, i+1)
    subplot.title.set_text(f'{filenames[i]}\n prediction: {results[i]}')
    subplot.imshow(Image.open(os.path.join(TEST_DATA_DIR, filenames[i])), cmap='gray')
    
plt.show()
```
